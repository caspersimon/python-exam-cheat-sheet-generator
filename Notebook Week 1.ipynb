### Objects have exactly one value<br>Depending on the type, objects are mutable [imutable], and this means that the value can [cannot] be changed 

```python
var_1 = 100
id1 = id(var_1)
var_1 = 110
id2 = id(var_1)
print(id1 == id2)

var_1 = 100.0
id1 = id(var_1)
var_1 = 110.0
id2 = id(var_1)
print(id1 == id2)

var_1 = '100'
id1 = id(var_1)
var_1 = '110'
id2 = id(var_1)
print(id1 == id2)

var_1 = True
id1 = id(var_1)
var_1 = False
id2 = id(var_1)
print(id1 == id2)

var_1 = [100,110]
id1 = id(var_1)
var_1[0] = 120
var_1.append(130)
id2 = id(var_1)
print(id1 == id2)

var_1 = {'first': 100,'second': '110'}
id1 = id(var_1)
var_1['first'] = 120
id2 = id(var_1)
print(id1 == id2)

var_1 = {1, '2', 3}
id1 = id(var_1)
var_1.remove(3)
var_1.add(4)
id2 = id(var_1)
print(id1 == id2)
```

### Tupples are also immutable, changing a tupple gives an error

```python
var_1 = (100, '100')
id1 = id(var_1)
var_1[1] = '110'
id2 = id(var_1)
print(id1 == id2)
```

### The cleaner way to deal with errors in Python

```python
var_1 = (100, '100')
id1 = id(var_1)
try:
    var_1[1] = '110'
except:
    print ('tupples cannot be changed')
id2 = id(var_1)
print(id1 == id2)
```

### Importing names of objects for using extra functionality

```python
# In this cell we create a function that deletes all the thus far created names in your program, 
# we need such a function else you cannot run some cells more times and getting the same result all the time when testing the import statements 
import sys
def clear_names():
    message = ''
    for n in sorted(globals()): 
        if n[0]!='_' and n not in ["sys", "clear_names"]:
            message += f"{n}, "
            delattr(sys.modules[__name__], n)
```
