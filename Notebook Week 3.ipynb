### Accepting positional arguments

```python
def adder(n1, n2):
    return n1 + n2
print(adder(1, 2))    
```

```python
def adder(*args):
    result = 0
    for number in args:
        result += number  
    return result
print(adder(1, 2, 3, 4, 5))
    
```

```python
def adder(l1):
    result = 0
    for number in l1:
        result += number  
    return result
print(adder([1, 2, 3, 4, 5]))
```

```python
def adder(n1, n2, *args):
    result = n1 + n2
    for number in args:
        result += number  
    return result
print(adder(1, 2, 3, 4, 5))
```

### A problems with mutable arguments

```python
def adder(l2):
    result = 0
    while l2:
        result += l2[0]
        del l2[0]  
    return result
l1 = [1, 2, 3, 4, 5]
total = adder(l1)
print(f"The sum of {l1} = {total}")
```

```python
def adder(l2):
    l2 = l2[:]
    result = 0
    while l2:
        result += l2[0]
        del l2[0]  
    return result
l1 = [1, 2, 3, 4, 5]
total = adder(l1)
print(f"The sum of {l1} = {total}")
```

### Function factories, functions creating other functions

### Accepting keyword arguments

```python
def calc(first, second, third, fourth, fifth, sixt):
    return first + 2 * second + 3 * third + 4 * fourth + 5 * fifth + 6 * sixt 
print(calc(fifth = 2, third=4, fourth=1, sixt=5, second=2, first=3))   
```

```python
def calc(**kwargs):
    return kwargs['first'] + 2 * kwargs['second']+ 3 * kwargs['third'] + 4 * kwargs['fourth'] + 5 * kwargs['fifth']+ 6 * kwargs['sixt']
print(calc(third=4, fourth=1, second=2, first=3, fifth = 2, sixt=5))
    
```

### You can mix positional and keyword arguments

```python
def calc(first, second, *args, **kwargs):
    return first + second * 2 + args[0] * 3 + args[1] * 4 + 5 * kwargs['fifth']+ 6 * kwargs['sixt']
print(calc(3, 2, 4, 1, sixt=5, fifth=2))
    
```

### Lambda functions can be used in a style of programming called functional programming<br>First function: the map function 

```python
l1 = [1, 2, 3, 4, 5, 6]
l2 = []
for item in l1:
    l2 += [item * 2]
print(l2)    
```

```python
l1 = [1, 2, 3, 4, 5, 6]
print(list(map(lambda x: x * 2, l1)))
```

```python
l1 = ['aaa', 'baab', 'c']
print(list(map(lambda x: x.count('a'), l1)))
```

### Second function: the filter function 

```python
l1 = [1, 2, 3, 4, 5, 6]
l2 = []
for item in l1:
    if not item % 2:
        l2 += [item]
print(l2)    
```

```python
l1 = [1, 2, 3, 4, 5, 6]
print(list(filter(lambda x: not x % 2, l1)))
```

```python
l1 = ['aaa', 'baab', 'c']
print(list(filter(lambda x: x.count('a') > 0, l1)))
```

### Third function: the reduce function 

```python
l1 = [1, 2, 3, 4, 5, 6]
result = l1[0]
for item in l1[1:]:
    result *= item
print(result)    
```

```python
from functools import reduce
l1 = [1, 2, 3, 4, 5, 6]
print(reduce(lambda x, y: x*y, l1))
```

### The reduce function can have a third parameter, if this is the case this will be used to initialize the result instead of the first element of the sequence 

```python
l1 = ['aaa', 'baab', 'c']
result = 0
for item in l1:
    result += item.count('a')
print(result) 
```

```python
from functools import reduce
l1 = ['aaa', 'baab', 'c']
print(reduce(lambda x, y: x + y.count('a'), l1, 0)) 
```

### A while loop can do the same as a for-loop. Comparing them can make you more aware what a for-loop actually does

```python
x = [5, 10, 21]
result = []
for el in x:
    if el != 10:
        result.append(el * 2 if el < 10 else el // 2)
print(result)
```

```python
x = [5, 10, 21]
result = []
index = 0
while index < len(x):
    el = x[index]
    if el != 10:
        result += [el * 2 if el < 10 else el // 2]
    index += 1
print(result)
```

### Nested for-loops

```python
x = ['a', 'b', 'c', 'd']
y = ['e', 'f']
z = ['g', 'h']
result = []
for elx in x:
    for ely in y:
        for elz in z:
            result += [elx + ely + elz]
print(result)
```
